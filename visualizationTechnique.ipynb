```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Car data.csv' )               # reading the csv file
df.head()                                       # displaying the top 5 rows
```

```python
# Scatter plot for Selling Price vs Present Price
plt.figure(figsize=(8, 6))
plt.scatter(df['Present_Price'], df['Selling_Price'], color='blue')
plt.title('Selling Price vs Present Price')
plt.xlabel('Present Price')
plt.ylabel('Selling Price')
plt.grid(True)
plt.show()
```

```python
# Histogram for Present Price distribution 
# Extract Present_Price data
present_prices = df['Present_Price']
# Create histogram
plt.hist(present_prices, bins=5, color='salmon', edgecolor='black')

# Add labels and title
plt.xlabel('Present Price')
plt.ylabel('Frequency')
plt.title('Histogram of Present Price')

# Show plot
plt.show()
```

```python
# Heatmap for correlation matrix 
# Calculate the correlation matrix
corr = df.corr()

# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Bar chart for 'Fuel_Type' column
df['Fuel_Type'].value_counts().plot(kind='bar', figsize=(8, 5), color='skyblue')
plt.title("Fuel Type Distribution")
plt.ylabel('Number of Occurrences')
plt.xlabel('Fuel Type')
plt.xticks(rotation=0)  # Rotate x-axis labels if needed
plt.show()
```

```python
fuel_types = df['Fuel_Type'].value_counts()

# Pie chart for 'Fuel_Type' column
plt.figure(figsize=(8, 6))
plt.pie(fuel_types, labels=fuel_types.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Fuel Types')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
