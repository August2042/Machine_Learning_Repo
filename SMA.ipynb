```python
#importing the libraries
import pandas as pd
import numpy as np
import seaborn as sns                       #visualisation
import matplotlib.pyplot as plt             #visualisation
%matplotlib inline     
sns.set(color_codes=True)
```

```python
df = pd.read_csv('sentimentdataset.csv')   #importing the dataset
df.head(5)                                 #displaying first five rows
```

```python
df.tail(5)                          #displaying last five rows
```

```python
df.dtypes                       #checking the data type
```

```python
df = df.drop(['Unnamed: 0.1', 'Unnamed: 0'], axis=1)            #dropping the unnecessary columns
df.head(5)
```

```python
# Rename the columns
df = df.rename(columns={
    'Text': 'Post_Content',
    'Sentiment': 'Sentiment_Label',
    'Timestamp': 'Post_Time',
    'User': 'Username',
    'Platform': 'Social_Media_Platform',
    'Hashtags': 'Post_Hashtags',
    'Retweets': 'Num_Retweets',
    'Likes': 'Num_Likes',
    'Country': 'User_Country',
    'Year': 'Post_Year',
    'Month': 'Post_Month',
    'Day': 'Post_Day',
    'Hour': 'Post_Hour'
})

# Display the first 5 rows with the renamed columns
df.head(5)
```

```python
df.shape                   #used to know the number of rows and columns
```

```python
duplicate_rows_df = df[df.duplicated()]                 #used to know the duplicate rows
print("number of duplicate rows: ", duplicate_rows_df.shape)
```

```python
df.count()      # Used to count the number of rows
```

```python
df = df.drop_duplicates()              #used to remove the duplicate rows
df.head(5)
```

```python
df.count()                      # Used to count the number of rows
```

```python
print(df.isnull().sum()) #Check the missing values in the dataset
```

```python
#Boxplot
sns.boxplot(x=df['Num_Retweets'])


```

```python
#Boxplot
sns.boxplot(x=df['Num_Likes'])


```

```python
# Bar chart for 'User' column
df['Username'].value_counts().nlargest(40).plot(kind='bar', figsize=(10, 5), color='skyblue')
plt.title("User Distribution in Sentiment Dataset")
plt.ylabel('Number of Occurrences')
plt.xlabel('Users')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.show()
```

```python
# Bar chart for user country distribution
df['User_Country'].value_counts().nlargest(15).plot(kind='bar', figsize=(10, 5), color='skyblue')
plt.title("User Country Distribution")
plt.ylabel('Number of Users')
plt.xlabel('Countries')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Bar chart for platform-wise counts
df['Social_Media_Platform'].value_counts().plot(kind='bar', figsize=(10, 5), color='skyblue')
plt.title("Distribution Across Social Media Platforms")
plt.ylabel('Number of Posts')
plt.xlabel('Social Media Platforms')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Bar chart for post distribution by hour
df['Post_Hour'].value_counts().sort_index().plot(kind='bar', figsize=(10, 5), color='skyblue')
plt.title("Hourly Distribution of Posts")
plt.ylabel('Number of Posts')
plt.xlabel('Hour of the Day')
plt.show()
```

```python
# Scatter plot for Num_Retweets and Num_Likes
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Num_Retweets', y='Num_Likes', data=df, color='skyblue')
plt.title('Scatter Plot: Num_Retweets vs. Num_Likes')
plt.xlabel('Number of Retweets')
plt.ylabel('Number of Likes')
plt.show()
```
